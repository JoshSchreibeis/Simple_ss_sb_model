```python
events = [
    {"user_id": 1, "item": "apple"},
    {"user_id": 2, "item": "banana"},
    {"user_id": 1, "item": "orange"},
    {"user_id": 2, "item": "apple"},
    {"user_id": 3, "item": "banana"},
    {"user_id": 4, "item": "chainsaw"},
]


def count_user_purchases(events):
    """
    events: list of dicts, each dict has "user_id" and "item"
    returns: dict mapping user_id -> purchase count
    """
    
    user_purchase_count = {}
    for event in events:
        user_id = event["user_id"]
        item = event["item"]
        user_purchase_count.setdefault(user_id,set()).add(item)
        #print(user_purchase_count)
    output = {user_id: len(item) for user_id,item in user_purchase_count.items()}
    
    return output
    
count_user_purchases(events)
```

```python
import numpy as np

data = [
    {"value": 10},
    {"value": 20},
    {"value": 30},
    {"value": 40},
]
store = []
for d in data:
    val = d["value"] 
    store.append(val)
mean_val = np.mean(store)
print(mean_val)
    

def mean_value(data):
    """
    data: list of dicts with key "value"
    returns: float mean of all values
    """
    total = 0 
    count = 0 

    for d in data:
        total += d["value"]
        count += 1
        
    return total/count 
mean_value(data)

```

```python
data = [
    {"value": 10},
    {"value": None},
    {"value": 30},
    {"value": None},
    {"value": 50},
]

def count_missing(data):
    """
    data: list of dicts with key "value"
    returns: int count of missing values
    """
    count = 0 
    for d in data:
        if d["value"] == None:
            count += 1 
    return count 
count_missing(data)
```

```python
data = [
    {"user_id": 1, "value": 10},
    {"user_id": 1, "value": None},
    {"user_id": 2, "value": 20},
    {"user_id": 2, "value": 30},
    {"user_id": 3, "value": None},
]

user_val_mean = {}
for d in data:
    id = d["user_id"]
    val = d["value"]
    if val != None:
        user_val_mean.setdefault(id,[]).append(val)

mean_per_user = {id: sum(val)/len(val) for id, val in user_val_mean.items()}
print(mean_per_user)

```

```python
events = [
    {"user_id": 1, "timestamp": 100},
    {"user_id": 2, "timestamp": 101},
    {"user_id": 1, "timestamp": 102},
    {"user_id": 3, "timestamp": 109},
    {"user_id": 2, "timestamp": 104},
    {"user_id": 1, "timestamp": 105},
]

K = 2


def top_k_active_users(events, K):
    uid_counts = {}
    for event in events:
        uid = event["user_id"]
        uid_counts[uid]= uid_counts.get(uid,0) +1
    top_k = sorted(uid_counts.items(), key=lambda x: x[1], reverse=True)[:K]
   
    return top_k
top_k_active_users(events,3)
```

```python
# Accuracy = (TP + TN) / total

# Precision = TP / (TP + FP)

# Recall = TP / (TP + FN)

y_true = [1, 0, 1, 1, 0, 0, 1]
y_pred = [1, 0, 0, 1, 0, 1, 1]

def stattest(y_true,y_pred):

    TP = 0
    TN = 0 
    FP = 0 
    FN = 0 
    
    for i in range(len(y_true)):
        if y_true[i] == y_pred[i]:
            if y_true[i] == 1:
                TP += 1
            else:
                TN += 1
        else:
            if y_true[i] == 0:
                FN += 1
                
            else:
                FP += 1 
                
        total = len(y_true)
        Accuracy = (TP + TN) / total

        Precision = TP / (TP + FP)

        Recall = TP / (TP + FN) 

    return print(Accuracy,Precision,Recall)
stattest(y_true,y_pred)

```
